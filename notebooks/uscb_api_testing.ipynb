```python
import requests, os
from dotenv import load_dotenv
import pandas as pd
load_dotenv()

resp = requests.get(
    "https://api.census.gov/data/timeseries/eits/mrts",
    params={
        "get": "cell_value,time_slot_id,category_code,seasonally_adj,data_type_code",
        "time": "2023-01",          # single month — minimal response
        "key": os.environ["USCB_API_KEY"],
    }
)

headers, *rows = resp.json()
```

```python
df = pd.DataFrame(rows, columns=headers)
df = df[df["category_code"] == "44000"]
print(df[["category_code", "data_type_code", "seasonally_adj", "time_slot_id"]].drop_duplicates())
```
